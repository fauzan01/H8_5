```python
#import library yang di butuhkan dan loading dataset nya
import pandas as pd
import scipy.stats as stats
import numpy as np

file = 'nyc-rolling-sales.csv'
df = pd.read_csv(file)
df
```

# preprocessing 

```python
df['SALE PRICE'].unique()
```

```python
#menghapus baris yang isi nya strip di kolom SALE PRICE
df['SALE PRICE'] = df['SALE PRICE'].replace(' -  ', float('nan'))
df = df.dropna(subset=['SALE PRICE'])

#convert ke numerik
df['SALE PRICE'] = pd.to_numeric(df['SALE PRICE'], errors='coerce')
```

```python
#melihat mean, standar deviasi, range dll
df['SALE PRICE'].describe()
```

```python
#mencari nilai median dan modus
print("modus = ", df['SALE PRICE'].mode())
print("median = ",df['SALE PRICE'].median())
```

Uji hipotesis menggunakan ANOVA 

```python
df['BOROUGH'].unique()
```

```python
from scipy import stats

group1 = df[df['BOROUGH'] == 1]['SALE PRICE']
group2 = df[df['BOROUGH'] == 2]['SALE PRICE']
group3 = df[df['BOROUGH'] == 3]['SALE PRICE']
group4 = df[df['BOROUGH'] == 4]['SALE PRICE']
group5 = df[df['BOROUGH'] == 5]['SALE PRICE']

# Lakukan uji ANOVA
statistic, p_value = stats.f_oneway(group1, group2, group3, group4, group5)

# Tampilkan hasil
print("Statistic:", statistic)
print("P-value:", p_value)
```
