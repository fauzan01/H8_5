```python
%matplotlib inline
import pandas as pd
import mplfinance as mpf


file = 'GBPJPY_historical_data.csv'
data = pd.read_csv(file)

data['Date'] = pd.to_datetime(data['Date'])  # Convert the 'Date' column to datetime format
data['Time'] = data['Date'].dt.time  # Extract the time from the 'Date' column
data['Dateonly'] = data['Date'].dt.date

data.Dateonly = pd.to_datetime(data.Dateonly)

data.drop(["Unnamed: 7","Change(Pips)"], axis = 1, inplace = True)
data.drop(["Change(%)"], axis = 1, inplace = True)

data
```

```python
data['Date'] = pd.to_datetime(data['Date'])  # Convert the 'Date' column to datetime format
data['Time'] = data['Date'].dt.time  # Extract the time from the 'Date' column
data['Dateonly'] = data['Date'].dt.date

# Filter data for the specific date (e.g., '2023-05-23')
specific_date = '2023-05-23'
filtered_data = data[data['Dateonly'] == pd.to_datetime(specific_date).date()]
x=filtered_data.iloc[::-1]
x.index = pd.DatetimeIndex(filtered_data['Date'])
mpf.plot(x,type='candle')
```
