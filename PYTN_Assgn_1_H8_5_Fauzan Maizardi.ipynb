Hipotesis Statistik
1. Apa jenis kasus kriminal yang paling banyak pernah terjadi dari tahun ke tahun?
2. Dimana kasus kriminal tersebut biasanya terjadi?
3. bagaimana perkembangan kasus kriminal tersebut?

```python
#memasukkan library yang dibutuhkan untuk menggunakan dataset London Crime
import pandas as pd
file = 'london_crime_by_lsoa.csv'
df = pd.read_csv(file)
df
```

```python
#mencari Kasus kriminal apa yang paling banyak terjadi dari tahun ke tahun?
df.nlargest(10,'value')
```

Untuk menjawab hipotesis pertama, yakni apa kasus kriminal yang paling banyak pernah terjadi dari sepanjang tahun, adalah kasus Possession Of Drugs. bisa dilihat langsung melalui table di atas, tanpa harus menggunakan visualisasi.

```python
#pencarian dimana kasus kriminal tersebut biasanya terjadi.
x = df[df['minor_category']=='Possession Of Drugs'] 
x1 = x[x['value']>0]
x1
```

```python
#mencari daerah mana yang paling banyak terjadi kasus Possession Of Drugs
x2 = x1['borough'].value_counts()
x2
```

```python
#memvisualisasikan daerah mana yang paling banyak terjadi kasus Possession Of Drugs
import matplotlib.pyplot as plt

x2.plot(kind='bar')

plt.xlabel('Unique Elements')
plt.ylabel('Frequency')
plt.title('Perbandingan Banyak Kasus Possession Of Drugs per Daerah')
plt.show()
```

Dari visualisasi Bar Chart di atas, bisa dilihat kasus Possession Of Drugs lebih banyak terjadi di Daerah Brent, Southwark, dan Lambeth. 

```python
#memvisualisasikan bagaimana perkembangan kasus kriminal tersebut dari tahun ke tahun
x1.groupby('year')['value'].sum().plot(kind='line')
plt.xlabel('year')
plt.ylabel('Total Value')
plt.title('Total Value per Year')
plt.show()
```

Dari visualisasi Bar Chart di atas, bisa dilihat kasus Possession Of Drugs dari tahun 2008 ke tahun 2016 menurun hampir setengah dari kasus yang terjadi sejak tahun 2008.
