# A simple example of linear regression concept, if i had amnesia 

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Contoh data
x = [[1], [2], [3], [4], [5], [6], [7], [8]]
y = [1,3,5,7,9,11,13,15]  #ini angka nya sengaja dibikin rapi aja, maka di hasil prediksi nya rapi juga

# Membuat model regresi linier
model = LinearRegression()

# Melatih model
model.fit(x, y)

# Mendapatkan nilai kemiringan (m) dan perpotongan (c)
slope = model.coef_[0]  # m
intercept = model.intercept_  # c

print("Slope (m):", slope)
print("Intercept (c):", intercept)
```

```python
y_pred = model.predict(x)

# Visualisasi data dan model
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x, y_pred, color='red', label='Regresi Linier')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#ini tuh poin2 y yg ada di garis merah di atas
print("Hasil Prediksi:", y_pred)
```

```python
from sklearn.metrics import r2_score

# Menghitung R-squared
r_squared = r2_score(y, y_pred)

print("R-squared (R²):", r_squared)

#jika hasil nya semakin mendekati angka 1, maka smakin akurat
```
