```python
import pandas as pd
```

```python
file = 'melb_data.csv'
df = pd.read_csv(file)
df
```

```python
df[['Rooms', 'Landsize', 'Distance','Price']].describe()
```

```python
#nyari jumlah room pada rumah yang harganya paling mehol
df[df['Price']== 9000000.0]
```

```python
#nyari berapa rata-rata harga rumah yang jarak ke CBD (variable Distance) terdekat?
df[df['Distance']==df['Distance'].min()]['Price'].mean()
```

```python
#nyari rentang harga rumah di Southern Metropolitan
house_in_southern_metropolitan = df[df['Regionname']=='Southern Metropolitan']
house_in_southern_metropolitan['Price'].max() - house_in_southern_metropolitan['Price'].min()
```
